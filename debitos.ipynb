```python
# Importar tabela de controle do excel
import pandas as pd

controle_df = pd.read_excel("Controle_test.xlsx")
display(controle_df)
```

```python
total = controle_df.iloc[0,3]
saldo = controle_df.iloc[0,4]
print(total)
print(saldo)
```

```python
# Adicionar novos débitos a tabela
novos_debitos = []
conta = 's'

while conta != '':
    conta = input("Conta: ")
    if conta != '':
        valor = float(input("Valor: "))
        novos_debitos.append([conta, valor])
        total += valor
        saldo -= valor
print(novos_debitos)
```

```python
# Converter novos debitos em dataframe
nv_df = pd.DataFrame(novos_debitos, columns=["Conta:", "Valor:"])
display(nv_df)
```

```python
# Concatenar dataframes
controle_df = pd.concat([controle_df, nv_df], ignore_index= True)
display(controle_df)
```

```python
# Atualizar Total
controle_df.loc[0, "Gasto total:"] = total
# Atualizar saldo
controle_df.loc[0,"Saldo:"] = saldo
display(controle_df)
```

```python
# Exportar exel

controle_df.to_excel("Controle_test.xlsx", index=False )
```
