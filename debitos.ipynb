```python
import pandas as pd

controle_df = pd.read_excel("Controle_test.xlsx")
display(controle_df)
```

```python
total = controle_df.iloc[0, 1]
saldo = controle_df.iloc[0, 2]
print(total, saldo)
```

```python
tabela = []
conta = 's'

while conta != '':
    conta = input("Conta: ")
    if conta != '':
        valor = float(input("Valor: "))
        tabela.append([conta, valor])
        total += valor
        saldo -= valor
print(tabela)
```

```python
print(total,saldo)
```

```python
# Converter novos tabela em dataframe
df = pd.DataFrame(tabela, columns=["Conta:", "Valor:"])
display(df)
```

```python
# Concatenar dataframes
controle_df = pd.concat([controle_df, df], ignore_index= True)
display(controle_df)
```

```python
# Atualizar Total
controle_df.iloc[0, 1] = total
# Atualizar saldo
controle_df.iloc[0, 2] = saldo
display(controle_df)
```

```python
# Atualizar Total
controle_df.loc[0, "Gasto Valor"] = total
# Atualizar saldo
controle_df.loc[0,"Saldo:"] = saldo
display(controle_df)
```

```python
# Exportar exel

controle_df.to_excel("Controle_test_novo.xlsx", index=False )
```
